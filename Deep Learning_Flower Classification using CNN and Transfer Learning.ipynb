```python
from keras import layers
from keras import models
```

```python
#Created sequential models using Keras
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(5, activation='sigmoid'))
model.summary()
```

```python
# from tensorflow import keras
from keras import optimizers
model.compile(optimizer='Adam', 
              loss='categorical_crossentropy', 
              metrics =['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(flower_train_dir,
                                                    target_size=(224, 224), 
                                                    batch_size=10,
                                                    class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(flower_validation_dir,
                                                        target_size=(224, 224),
                                                        batch_size=10,
                                                        class_mode='categorical')
```

```python
history = model.fit_generator(train_generator,
                              steps_per_epoch=100,
                              epochs=10,
                              validation_data=validation_generator,
                              validation_steps=10)
```

```python
model.save('flower_classification.h5')

import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy on CNN')
plt.xlabel('Epoch')
#plt.ylabel('your ylabel')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

**Tranfer Learning: VGG16 Model**

```python
from keras.layers import pooling
from keras.layers import Activation, Dense
from keras.applications.vgg16 import VGG16, preprocess_input
```

```python
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(240, 240, 3))
```

```python
conv_base.summary()
```

```python
num_classes = 5
image_size = 224
batch_size_training = 30
batch_size_validation  = 20
```

```python
data_generator = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 20,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip = True,
    width_shift_range=0.1,
    height_shift_range=0.1,
    preprocessing_function=preprocess_input)
```

```python
Train_generator = data_generator.flow_from_directory(
    "/content/small_dataset/train",
    target_size=(image_size,image_size),
    batch_size = batch_size_training,
    class_mode = "categorical")

Validation_generator = data_generator.flow_from_directory(
    "/content/small_dataset/validation",
    target_size=(image_size,image_size),
    batch_size = batch_size_training,
    class_mode = "categorical")
```

```python
modelvgg16 = models.Sequential()
```

```python
modelvgg16.add(VGG16(include_top=False, pooling="avg", weights="imagenet"))
```

```python
modelvgg16.add(Dense(num_classes, activation="softmax"))
modelvgg16.layers
```

```python
modelvgg16.layers[0].layers
```

```python
modelvgg16.layers[0].trainables=False
modelvgg16.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
steps_per_epoch_train = len(Train_generator)
steps_per_epoch_valid = len(Validation_generator)
number_epochs = 90
```

```python
fit_history = modelvgg16.fit_generator(Train_generator,
                                       steps_per_epoch=steps_per_epoch_train,
                                       epochs=number_epochs,
                                       verbose=1,
                                       validation_data=Validation_generator,
                                       validation_steps=steps_per_epoch_valid)
```

```python
modelvgg16.save('flower_class.h5')

import matplotlib.pyplot as plt

acc = fit_history.history['accuracy']
val_acc = fit_history.history['val_accuracy']
loss = fit_history.history['loss']
val_loss = fit_history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy VGG16')
plt.xlabel('Epoch')
#plt.ylabel('your ylabel')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss on VGG16')
plt.xlabel('Epoch')
#plt.ylabel('your ylabel')
plt.legend()

plt.show()
```
